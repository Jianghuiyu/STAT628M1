### 3. Variable selection

#### 3.1 Stepwise method

##### 3.1.1 Using AIC  to select variables.

<center>$lnL(\mu,\sigma^2)=-\frac{n}{2}ln(2\pi\sigma^2)-\frac{1}{2\sigma^2}\sum_{i=1}^n(x_i-\mu)^2$</center>

<center>AIC: $2k-2ln(\hat{L})$</center>

According to the AIC, we pick five variables include abdomen, weight, wrist, biceps, and age.We can get the model,

<center>$(Body Fat\%)=-23.717+0.838Abdomen-0.085Weight-1.656Wrist+0.039Age+0.277Biceps ;\ \ Multiple R^{2}:0.7356$</center>

Since the BIC is more conservative, we use the stepwise model based on BIC criterion.



##### 3.1.2 Using BIC to select variables.

<center>BIC: $ln(n)k-2ln(\hat{L})$</center>

The result of BIC shows abdomen, weight, and wrist are the most important variables for interpret the bodyfat.We can get the model,

<center>$(Body Fat\%)=-23.994+0.885Abdomen-0.087Weight-1.282Wrist;\ \ Multiple R^{2}:0.7292$</center>

#### 3.2 Elastic net

<center>$\hat{\beta}=argmin_{\beta}(||y-X\beta||^2+\lambda_2||\beta||^2+\lambda_1||\beta||_1)$</center>

![Image](https://github.com/Jianghuiyu/STAT628M1/blob/master/Image/lassofit.png?raw=true)

We use the grid search to determine the alpha based on the regularization path plot. Next, cross validation is adopted to choose the proper lambda. According to the graphs about the regularization path and Cross-validation, we can get the final model selects the age, height, adiposity, abdomen, and wrist. We can get the model,

<center>$(Body Fat\%)=5.717+0.736Abdomen-0.336Height-1.692Wrist+0.045Age-0.091 ;\ \ Multiple R^{2}:0.7341$</center>


#### 3.3 Which one to choose?

For the sake of simplicity, we choose the stepwise model based on BIC criterion as our final model. First of all, the performance of the regressions does not differ a lot between each model. Each model’s R square are very close. Both AIC and elastic net pick too many variables. Secondly, it is strange that some estimations in the regularization path of elastic net change the sign as the penalty grows. In addition, the cross-validation only care about the mse. However, we also want to make some inference in our model and in that case, cross validation might not very suitable to use.

### 4.Model fitting

The final model is 
<center>$(Body Fat\%)=-23.994+0.885Abdomen(cm)-0.087Weight(lbs)-1.282Wrist(cm)$</center>

For the sake of simplicity, we simple our linear model as,
<center>$(Body Fat\%)=-24+0.9Abdomen(cm)-0.1Weight(lbs)-1.3Wrist(cm)$</center>

Interpretation: e.g. If we fix the Weight and Wrist, the Abdomen increases 1 lbs, the body fat will increase 0.9%

<center>$Multiple R^{2}:0.7292; Residual standard error: 3.981; p-value: < 2.2e-16$</center>
